```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
def show_training_graphs(graph_folder):
    important_graphs = [
        'confusion_matrix_normalized.png',
        'results.png',
        'F1_curve.png',
        'PR_curve.png',  # Precision-Recall curve
        'P_curve.png',   # Precision curve
        'R_curve.png',   # Recall curve
    ]
    
    
    titles = [
        'Confusion Matri Normalized',
        'Results',
        'F1 Curve',
        'Precision-Recall Curve',
        'Precision Curve',
        'Recall Curve',
    ]
    
    #
    plt.figure(figsize=(20, 20))
    
    plot_index = 1
    
    for graph_name, title in zip(important_graphs, titles):
        graph_path = os.path.join(graph_folder, graph_name)
    
        if os.path.exists(graph_path):
            img = cv2.imread(graph_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #
            
          
            plt.subplot(3, 2, plot_index)
            plt.imshow(img)
            plt.title(title, fontsize=16)
            plt.axis('off')
            
            plot_index += 1
    
    plt.tight_layout()
    plt.show()
```

```python
# Model with 110 epoch and yolov8s
graph_folder = 'runs\detect\\train7'  
show_training_graphs(graph_folder)
```

```python
# Model with 30 epochs and yolov8m
graph_folder = 'runs\detect\\train'  
show_training_graphs(graph_folder)
```

```python
# Model with 100 epochs and yolov8n
graph_folder = 'runs\detect\\train6'  
show_training_graphs(graph_folder)
```
